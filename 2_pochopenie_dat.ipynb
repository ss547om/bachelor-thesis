```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
data = pd.read_csv("data_sk.csv", sep = ",")
```

```python
pd.set_option('display.max_columns', 40)
data.head(3)
```

```python
poradie = ['Pondelok', 'Utorok', 'Streda', 'Stvrtok', 'Piatok', 'Sobota', 'Nedela']
sns.countplot(x = 'den_v_tyzdni', order = poradie, data = data)

plt.xlabel('Deň v týždni')
plt.ylabel('Počet dopravných nehôd')

plt.show()
```

```python
top_10_dialnice = data['cislo_dialnice'].value_counts().nlargest(10)

sns.barplot(x=top_10_dialnice.index, y=top_10_dialnice.values)

plt.xlabel('Číslo diaľnice')
plt.ylabel('Počet nehôd')

plt.show()
```

```python
top_10_pricin = data['pricina_nehody'].value_counts().nlargest(10)

sns.barplot(y=top_10_pricin.index, x=top_10_pricin.values)

plt.ylabel('Príčina dopravnej nehody')
plt.xlabel('Počet dopravných nehôd')

plt.show()
```

```python
sns.countplot(x = 'klasifikacia_nehody', data = data)

plt.xlabel('Klasifikácia nehody')
plt.ylabel('Počet dopravných nehôd')
plt.xticks(rotation=10)

plt.show()
```

```python
sns.countplot(x = 'faza_dna', data = data)

plt.xlabel('Fáza dňa')
plt.ylabel('Počet dopravných nehôd')

plt.show()
```

```python
sns.countplot(x = 'pocasie', data = data)

plt.xlabel('Počasie')
plt.ylabel('Počet dopravných nehôd')
plt.xticks(rotation=30)

plt.show()
```

```python
sns.boxplot(x= 'pocet_osob', data = data)

plt.xlabel('Počet osôb')

plt.show()
```
